```python
import requests
import pandas as pd
from dotenv import load_dotenv
import os
import json

load_dotenv()
API_KEY = os.getenv("EIA_API_KEY")

# Test endpoint retail sales
url = "https://api.eia.gov/v2/electricity/retail-sales"
params = {
    "api_key": API_KEY,
    "frequency": "monthly",
    "data[]": "sales",
    "facets[stateid][]": "US",
    "start": "2022-01",
    "end": "2022-03",  # ambil 3 bulan dulu untuk test
    "length": 10
}

resp = requests.get(url, params=params)
raw = resp.json()

# Lihat struktur response mentah
print("Status:", resp.status_code)
print("Keys di response:", raw.keys())
print()
print(json.dumps(raw, indent=2)[:3000])  # print 3000 karakter pertama
```

```python
url2 = "https://api.eia.gov/v2/electricity/electric-power-operational-data"
params2 = {
    "api_key": API_KEY,
    "frequency": "monthly",
    "data[]": "generation",
    "facets[location][]": "US",
    "facets[fueltypeid][]": "NG",  # natural gas dulu satu jenis
    "start": "2022-01",
    "end": "2022-03",
    "length": 10
}

resp2 = requests.get(url2, params=params2)
raw2 = resp2.json()

print("Status:", resp2.status_code)
print()
print(json.dumps(raw2, indent=2)[:3000])
```
